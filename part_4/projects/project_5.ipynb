```python
from enum import Enum
```

```python
class AppException(Enum):

    NotAnInteger = 100, ValueError, 'Value is not an integer'
    NotAString = 101, ValueError, 'Value is not a string'

    NoConnection = 200, RuntimeError, 'Cannot connect to a database'
    
    def __new__(cls, member_value, member_exception, member_message):
        member = object.__new__(cls)

        member._value_ = member_value
        member.error = member_exception
        member.message = member_message

        return member



    @property
    def code(self):
        return self.value
    

    def throw(self, custom_message=None):
        if custom_message is None:
            raise self.error(self.message)
        if not isinstance(custom_message, str):
            raise TypeError('custom message has to be of a <str> type')
        raise self.error(custom_message)
```
