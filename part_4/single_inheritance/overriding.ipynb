```python
class Shape:
    def __init__(self, name):
        self.name = name

    def info(self):
        return f'Shape.info was called for Shape({self.name})'
    
    def extended_info(self):
        return f'Shape.extended_info called for Shape({self.name})'
    
class Polygon(Shape):
    def __init__(self, name):
        self.name = name

    def info(self):
        return f'Polygon.info was called for Polygon({self.name})'
```

```python
p = Polygon('square')
```

```python
p.info()
```

```python
p.extended_info()
```

```python
class Shape:
    def __init__(self, name):
        self.name = name

    def info(self):
        return f'Shape.info was called for Shape({self.name})'
    
    def extended_info(self):
        return f'Shape.extended_info called for Shape({self.name})', self.info()
    
class Polygon(Shape):
    def __init__(self, name):
        self.name = name

    def info(self):
        return f'Polygon.info was called for Polygon({self.name})'
```

```python
p = Polygon('square')
```

```python
p.info()
```

```python
p.extended_info()
```

```python
class Person:
    def __str__(self):
        return 'Person.__str__ called'

class Student(Person):
    def __repr__(self):
        return 'Student.__repr__ called'
```

```python
s = Student()
```

```python
str(s)
```

```python
repr(s)
```

```python
class Person:
    def __str__(self):
        print('Person.__str__ called')
        return self.__repr__()
    
    def __repr__(self):
        return str(hex(id(self)))

class Student(Person):
    def __repr__(self):
        return 'Student.__repr__ called'
```

```python
s = Student()
```

```python
repr(s)
```

```python
str(s)
```
